```python
import pandas as pd
import random
import math

speed_grid = 8
entropy_grid = 9
speed_range = 1
entropy_range = 1
repeat = 3

def getDataInfo(name, url_prefix, task_per_worker=18):
    data = []
    counter = 0
    orderList = []
    for speed_idx in range(speed_grid):
        for entropy_idx in range(entropy_grid):
            counter += 1
            orderList.append(counter)
            speed = speed_idx * speed_range / speed_grid + 0.01
            entropy = speed_idx * speed_range / entropy_grid + 0.01
            data.append({
                'index': counter,
                'speed': speed,
                'entropy': entropy,
                'url': f'{url_prefix}{counter:03}.gif',
                'name': name
            })
    meta_df = pd.DataFrame(data)
    total_n = speed_grid * entropy_grid
    
    batches = []
    random.shuffle(orderList)
    for idx in range(math.ceil(total_n / task_per_worker)):
        task = {'name': name, 'task0': int(total_n/2)}
        for img_idx in range(task_per_worker):
            task[f'task_{img_idx+1}'] = orderList[idx*task_per_worker + img_idx]
        task[f'task_{task_per_worker+1}'] = random.randint(1, total_n) 
        batches.append(task)
    batch_df = pd.DataFrame(batches)
    return meta_df, batch_df
```

```python
a, b = getDataInfo('test', 'https://www.test.com/')
```
